### What does this paper do?
- a structural life-cycle model estimate of wealth effect on consumption with endogeous labor supply;
- allowing for non-separable utility;
- intra-household dynamics;

### Findings
- life-cycle heterogeneity in response to wealth shocks, i.e. risky returns and house prices;
- rejection of separability;
- labor supply is endogenous, insurance role of labor supply, i.e., drop in house prices increases labor supply;
- intra-family differences in wealth effect on labor supply;

### Model 

#### Household problem
 - __2 state variables__: end of period risky asset holding and house wealth in previous period;
 - __4 control variables__: consumption, housing services, male and female labor supply;
 - __preference parameteres__: same-good Frisch elasticity and between-good Frisch elasticity.

\begin{eqnarray}
max \quad E_t(\sum_{s=0}^{T-t} \beta^s u(C_{i,t+s},L_{i,1,t+s},L_{i,2,t+s},H_{i,t+s}))
\end{eqnarray}


\begin{eqnarray}
\begin{split}
s.t. \quad & A_{i,t} = R^p_{i,t}A_{i,t-1}+L_{i,1,t}W_{i,1,t}+L_{i,2,t}W_{i,2,t} - C_{i,t} - I_{i,t} \\
& \underbrace{I_{i,t}}_{\text{House Investment}} = P^h_{i,t}H_{i,t} - P^h_{i,t}(1-\delta)H_{i,t-1} \\
& \underbrace{R^p_{i,t}}_{\text{Portfolio Return} } = \underbrace{R^b}_{\text{Risk-free Return}} + \alpha^s_{i,t-1}\underbrace{(R^s_{i,t}-R^b) }_{\text{Excess Return} }
\end{split}
\end{eqnarray}

#### Structure of exogenous shocks


##### House price: 
\begin{eqnarray}
\log P_{i,t} = \log P_{i,t-1} + \underbrace{X'_{i,t}\beta}_{\text{Predictable Component}} + \underbrace {\xi^h_{i,t}}_{\text{Price Shock}}+\underbrace{\zeta_{i,t}}_{\text{Measurement Error}}
\end{eqnarray}

##### Wage: 
\begin{eqnarray}
   \begin{split}
   & \log W_{i,j,t} = X'_{i,j,t}\beta^W_j + F_{i,j,t} + \underbrace{u_{i,j,t}}_{\text{i.i.d. Transitory Shock}} \\
   & F_{i,j,t} = F_{i,j,t-1} + \underbrace{v_{i,j,t}}_{\text{Permanent Shock}} \\
   \end{split}
\end{eqnarray}

##### Risky asset return:
\begin{eqnarray}
  r^s_{i,t} - r^b = \underbrace{X'_{i,t} \beta^s}_{\text{Predictable Component}} + \underbrace{\theta_i}_{\text{Household Fixed Effect} } + \underbrace{\xi^s_{i,t}}_{i.i.d. Shock}
\end{eqnarray}

##### Correlation:  
  - Positive correlation between permanent wage shocks and house price 
  - Positive correlation of permanent and transitory shocks between husband and wife

### Estimation 

#### Parameters to be estimated 

- joint moments of decision variables. $\Delta c_{i,t}$, $\Delta l_{i,j,t}$, $\Delta h_{i,t}$
- exogeneous shocks: $\xi^h$, $u_{i,j,t}$, $v_{i,j,t}$, $\xi^s_{i,t}$

- preference: $\eta_{c,p^h}$, $\eta_{l,p^h}$, $\eta_{c,r^s}$, $\eta_{l,r^s}$, etc.

### Discussion of the Paper
#### Model

- house as a non-durable good consumption, no discrete choice of house purchase. 
- therefore, only focuses on owners, this paper does not say anything about house-purchasing decision.

#### Estimation

- very likely biased in estimating human wealth.
- typically, the first step regression has small $R^2$ in income, portofolio share and risky returns.
- any small bias distorts projected human wealth substantially. 

\begin{eqnarray}
     \widehat {HumanWealth}_{i,j,t} = \sum^{T}_t \frac{\widehat Y_{i,j,t+k}}{(1+\widehat r^p_{i,t+k})^k}
\end{eqnarray}


\begin{eqnarray}
     \widehat r^p_{i,t+k} = r + (1-\widehat \alpha^s_{i,t+k}) \widehat r^s_{i,t+k}
\end{eqnarray}

#### Conclusion 

- Highlights the importance of endogenizing labor supply when considering wealth effects 
- extensive margin labor response is interesting although not using the model
- the linear approximation needs to be justified better
- estimation of human wealth and other variables are subject to potential bias
